```python
def file_read(file):
    data = pd.DataFrame()
    time = [] 
    camera = []
    direction = []
    number = []
    raw = []
    with open(file)  as f:
        f = f.readlines()
        for i in f:
            i = json.loads(i)
            camera.append(i.get('camera_id'))
            time.append(i.get('time'))
            direction.append(i.get('direction'))
            number.append(i.get('number'))
            raw.append(i.get('raw'))

    data['time'] = time
    data['camera'] = camera
    data['direction'] = direction
    data['number'] = number
    data['raw'] = raw
    return data
```

```python
paidui_data = pd.DataFrame()
all_file = os.listdir('/Users/panbingqing/Documents/工作/城市大脑/长沙调控/paidui')
all_file.sort()
all_file = all_file[1:]
#print(all_file)
for file in all_file:
    file = '/Users/panbingqing/Documents/工作/城市大脑/长沙调控/paidui'+ '/' + file
    
    print(file)
    #print(file)
    info = file_read(file)
    if len(info) == 0:
        continue
    paidui_data = paidui_data.append(info)    
```
